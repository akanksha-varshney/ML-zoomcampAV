```python
u = np.array([1,3,4,1])
v = np.array([3,4,0,1])
u.shape[0]
v.shape[0]
```

```python
def u_v_mult(u,v) : 
 u.shape[0]==v.shape[0]
 n = u.shape[0]
 result =0
 for i in range(n) : 
        result = result+ u[i]*v[i]
 return result  
    
```

```python
print(u_v_mult(u,v))
```

```python
U.shape[1]
```

```python
def U_v_mult(U,v) : 
 assert U.shape[1]==v.shape[0]
 rows = U.shape[0]
 result = np.zeros(rows)
 for i in range(rows):
        result[i] = u_v_mult(U[i],v)
 return result
```

```python
U_v_mult(U,v)
```

```python
def U_V_Multp(U,V):
 U.shape[1]==V.shape[0]
 rows = U.shape[1]
 cols = V.shape[0]
 result = np.zeros((rows,cols))
 for i in range(rows):
    for j in range(cols):
        result[i][j] = u_v_mult(U[i,:],V[:,j])
 return result
```

```python
U_V_Multp(U,V)
```
